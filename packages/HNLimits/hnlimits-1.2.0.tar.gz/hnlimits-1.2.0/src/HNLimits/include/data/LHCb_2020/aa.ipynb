```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
for i in range(1,5):
    print(i)
    data.append(np.genfromtxt(f"data_{i}.dat").T)
data = np.array(data)
```

```python
np.shape(data)
```

```python

for i in range(0,3):
    plt.plot(data[i,0],data[i,1], label='i -- first col')
    plt.plot(data[i,0],data[i,2], label='i -- second col',ls='--')

e,u = np.genfromtxt('LNV_corrected.dat', unpack=True)
plt.plot(e,u, lw=2, c='black')
e,u = np.genfromtxt('LNC_corrected.dat', unpack=True)
plt.plot(e,u, lw=2, c='black', ls='--')


plt.legend()
plt.yscale('log')
plt.xscale('log')
```
